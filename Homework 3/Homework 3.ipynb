# Generalization Error

The modified Hoeffding Inequality provides a way to characterize the generalization error with a probabilistic bound

$$
\mathbb{P}[|E_{in}(g)-E_{out}(g)|>\epsilon] \leq 2Me^{-2\epsilon^2N}
$$

for any $\epsilon > 0$.

## Question 1:

If we set $\epsilon = 0.05$ and want the probability bound $2Me^{-2\epsilon^2N}$ to be at most 0.03, what is the least number of examples $N$ needed for the case $M = 1$

$$
2Me^{-2\epsilon^2N} \leq 0.03 \implies 2Me^{-0.005N} \leq 0.03 \implies e^{-0.005N} \leq \frac{0.03}{2M} \implies N \geq \frac{\ln(\frac{0.03}{2M})}{-0.005}
$$

Setting $M=1$ we get: $N \geq 840$

Correct Answer: **B** (at least 1000)
